```python
plt.figure(figsize=(10,5))
sns.boxplot(x=df['release_year'])
sns.stripplot(x=df['release_year'], color='red', alpha=0.3)
plt.title("Boxplot with Outliers Highlighted")
plt.show()
```

```python
Q1 = df['release_year'].quantile(0.25)
Q3 = df['release_year'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers = df[(df['release_year'] < lower) | (df['release_year'] > upper)]
outliers
```

```python
plt.figure(figsize=(10,5))
sns.histplot(df['release_year'], kde=True, bins=30)
plt.title("Distribution (Bell Curve)")
plt.xlabel("Release Year")
plt.ylabel("Frequency")
plt.show()
```

# EDA Conclusion - Netflix Dataset

##Data Cleaning
- Removed **director** and **cast** columns due to high number of missing values and low importance for current analysis.

##Handling Missing Values
- Filled missing values in columns like **country, rating, duration** using **mode (most frequent value)**.
- Filled missing values in **date_added** using forward fill method.

##Outlier Analysis
- Checked for outliers using boxplot and IQR method.
- No significant outliers were found in the dataset.

##Overall Summary
- The dataset is now clean and ready for visualization and analysis.
